```python
import numpy as np
from matplotlib import pyplot as plt

```

```python
T = np.linspace(5,100,1000)
L = 2.897771955/T
plt.plot(T,L, color="lightseagreen")
plt.xlabel("Temperature (K)")
plt.ylabel("Peak's Wavelength (mm)")
plt.title("Wein's law")
plt.show()
```

```python
GHz = 1000000000
h = 6.62607015*(10^(-34)) #J/Hz
k = 1.380649*(10^(-23) ) #J/K
c = 299792458 #m/s
for t in [5,10,15,20,25,30,35,40]:

    f = np.linspace(30, 260, 1000000000)
    a = h*f*GHz/(k*t)
    e = np.exp(a)-1
    b = 1/e
    d = 2*h*((f*GHz)^3)/(c^2)
    B = d*b*(10^26)
    plt.plot(f,B,label="T = "+t+"K")
    plt.xlabel("Frequency (GHz)")
    plt.ylabel("Bn (Jy)")
    plt.legend()
    plt.show
```
